```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

import seaborn as sns
```

```python
mse = pd.read_csv('Train_new/mse_loss.csv', names=["MSE null checkpoint"])
mae = pd.read_csv('Train_new/val_mae_loss.csv', names=["MAE null checkpoint"])

mse_checkpoint = pd.read_csv('Train_from_places/mse_loss.csv', names=["MSE pre-trained checkpoint"])
mae_checkpoint = pd.read_csv('Train_from_places/val_mae_loss.csv', names=["MAE pre-trained checkpoint"])
```

```python
mse_merged = pd.concat([mse, mse_checkpoint], axis=1)

fig, axs = plt.subplots(1,2,figsize=(10,4))

sns.set_style("ticks")

sns.lineplot(data=mse_merged, ax=axs[0]).set(xlabel='Epochs', ylabel='MSE loss', xticks=np.arange(0,80,10))
sns.lineplot(data=mse_merged[10:], ax=axs[1]).set(xlabel='Epochs', title="Zoomed in", xticks=np.arange(0,80,10))

fig.suptitle("Training MSE loss with and without the base places365 checkpoint")
fig.tight_layout()

sns.despine()
plt.show()
fig.savefig("Combined MSE.png") 
```

```python
mae_merged = pd.concat([mae, mae_checkpoint], axis=1)

fig, axs = plt.subplots(1,1,figsize=(10,4))

sns.set_style("ticks")

sns.lineplot(data=mae_merged, ax=axs).set(xlabel='Epochs', ylabel='MAE loss', xticks=np.arange(0,80,10))

fig.suptitle("Training Valutation MAE Loss with and without the base places365 checkpoint")
fig.tight_layout()

sns.despine()
plt.show()
fig.savefig("Combined Val MAE.png") 
```

```python
directory = "train_gray/"

mse_gray = pd.read_csv(directory+'mse_loss.csv', names=["MSE gray"])
mae_gray = pd.read_csv(directory+'val_mae_loss.csv', names=["MAE gray"])
```

```python
sns.set_style("ticks")
plot = sns.lineplot(data=mse_gray)
plot.set(xlabel='Epochs', ylabel="MSE loss")
sns.despine()

fig = plot.get_figure()
plt.show()
fig.savefig("MSE_gray.png") 
```

```python
sns.set_style("ticks")
plot = sns.lineplot(data=mae_gray)
plot.set(xlabel='Epochs', ylabel='MAE loss')
sns.despine()

fig = plot.get_figure()
plt.show()
fig.savefig("Val MAE Gray.png") 
```

```python
mae_all = pd.concat([mae_merged, mae_gray], axis=1)
mse_all = pd.concat([mse_merged, mse_gray], axis=1)
```

```python
sns.set_style("ticks")
plot = sns.lineplot(data=mse_all)
plot.set(xlabel='Epochs', ylabel="MSE loss")
sns.despine()

fig = plot.get_figure()
plt.show()
fig.savefig("MSE_ALL.png") 
```

```python
sns.set_style("ticks")
plot = sns.lineplot(data=mae_all)
plot.set(xlabel='Epochs', ylabel="MSE loss")
sns.despine()

fig = plot.get_figure()
plt.show()
fig.savefig("MAE_ALL.png") 
```
